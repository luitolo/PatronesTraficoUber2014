# Clusterización mediante K-Means
Esta técnica no supervisada se basa en identificar grupos en los datos de tal manera que todos los datos del grupo (clúster) son datos con características similares mientras que los datos de los otros grupos son diferentes. El K-means es un método de Clustering que separa 'K' grupos de objetos (Clusters) de similar varianza, **minimizando un concepto conocido como inercia**, que es la suma de las distancias al cuadrado de cada objeto del Cluster a un punto 'μ' conocido como Centroide (punto medio de todos los objetos del Cluster).
El objetivo de K-Means es agrupar observaciones similares para descubrir patrones que a simple vista se desconocen. Para conseguirlo, el algoritmo busca un número fijo (k) de clústers en el dataset
El algoritmo intenta minimizar la distancia entre las observaciones que pertenecen a un clúster y su centroide. Es decir, el objetivo del K-Means es minimizar la suma de las distancias entre los puntos y el centroide al que pertenecen.

# Criterio para determinar el número de Cluster

1.	Los centroides dejan de cambiar. Es decir, después de múltiples iteraciones, los centroides de cada clúster no cambian de posición. Por lo que se asume que el algoritmo ha convergido.
2.	Los puntos dejan de cambiar de clúster. Parecido al anterior, pero esta vez no nos fijamos en los centroides, si no en los puntos que pertenecen a cada clúster. Cuando se observa que no hay un intercambio de clústers se asume que el modelo está entrenado.
3.	Límite de iteraciones. Podemos fijar un número máximo de iteraciones que queremos que nuestro algoritmo ejecute antes de pararlo. Cuando llega a ese número, se asume que el modelo no va a mejorar drásticamente y se para el entrenamiento.


```python
Sum_of_squared_distances = []

K = range(2,12)
for k in K:
    kmeans = KMeans(n_clusters=k,max_iter=300,random_state=24)
    kmeans.fit(clus)
    Sum_of_squared_distances.append(kmeans.inertia_)

# Encontrar el k optimo
plt.figure(figsize=(16,8))
plt.subplot(1,2,1)
plt.plot(K,Sum_of_squared_distances,'bx-')
plt.xlabel('Clusters k')
plt.ylabel('Sum_of_squared_distances')
plt.title('elbow method')
plt.grid()
plt.show()
```

Valor épsilon (eps) Epsilon es el radio del círculo alrededor de un punto de datos tal que todos los demás puntos de datos caen dentro del círculo se consideran puntos de vecindad. En otras palabras, dos puntos se consideran vecinos si la distancia entre ellos es menor o igual a eps. Si el valor de eps es extremadamente pequeño, entonces la mayoría de los puntos pueden no estar en el vecindario y se tratarán como valores atípicos. Esto dará lugar a una mala agrupación ya que la mayoría de los puntos no satisfacen el mínimo número de puntos necesarios para crear una región densa. Por otro lado, si se elige un valor extremadamente alto, la mayoría de los puntos de datos permanecerán en el mismo clúster. Esto volverá a crear una agrupación incorrecta donde varios clústeres pueden terminar fusionándose debido al alto valor de épsilon. Idealmente, debemos elegir el valor eps en función de la distancia del conjunto de datos (usando el gráfico k-distancia), sin embargo, normalmente se prefieren valores eps pequeños.

```python
dbscan = DBSCAN(eps= .025, min_samples = 8)
dbscan.fit(clus_DBScanScaled)
y_dbscan = dbscan.labels_
```


ML | V-Measure para evaluar el rendimiento de la agrupación en clústeres
Posted on julio 5, 2022 by Rudeus Greyrat	

Una de las principales desventajas de cualquier técnica de agrupamiento es que es difícil evaluar su desempeño. Para abordar este problema, se desarrolló la métrica de V-Measure .

El cálculo de la medida V primero requiere el cálculo de dos términos:-

Homogeneidad: una agrupación perfectamente homogénea es aquella en la que cada agrupación tiene puntos de datos que pertenecen a la misma etiqueta de clase. La homogeneidad describe la cercanía del algoritmo de agrupamiento a esta perfección.

Completitud: una agrupación perfectamente completa es aquella en la que todos los puntos de datos que pertenecen a la misma clase se agrupan en el mismo grupo. La completitud describe la cercanía del algoritmo de agrupamiento a esta perfección.

#Modelo de Serie de Tiempos

Una serie de tiempo es una secuencia en la que se registra una métrica en intervalos de tiempo regulares.
Cualquier serie de tiempo «no estacional» que muestre patrones y no sea un ruido blanco aleatorio se puede modelar con modelos ARIMA.
Si una serie de tiempo tiene patrones estacionales, entonces necesita agregar términos estacionales y se convierte en SARIMA, abreviatura de «Seasonal ARIMA». Más sobre eso una vez que terminemos ARIMA.
Porque, el término ‘Auto regresivo’ en ARIMA significa que es un modelo de regresión lineal que usa sus propios rezagos como predictores. Los modelos de regresión lineal, como sabe, funcionan mejor cuando los predictores no están correlacionados y son independientes entre sí.
Entonces, ¿cómo hacer estacionaria una serie?
El enfoque más común es diferenciarlo. Es decir, reste el valor anterior del valor actual. A veces, dependiendo de la complejidad de la serie, es posible que se necesite más de una diferenciación.
El valor de d, por lo tanto, es el número mínimo de diferenciaciones necesarias para hacer que la serie sea estacionaria. Y si la serie de tiempo ya es estacionaria, entonces d = 0.
